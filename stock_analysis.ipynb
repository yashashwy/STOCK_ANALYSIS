## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## LOADING THE DATASET AND DOING BASIC ANALYSIS


```python
df=pd.read_csv("Equity Data (1 Jan 2020).csv")
print(df.shape)
print(df.dtypes)
print(df.info)
```

## CONVERTING DATE1 INTO DATETIME OBJECT AND EXTRACTING MONTH AND YEAR AND PUTTING THEM IN A NEW COLUMN FOR EASE OF ANALYSIS

```python
df[' DATE1'] = pd.to_datetime(df[' DATE1'], dayfirst=True)
df['MONTH'] = pd.DatetimeIndex(df[' DATE1']).month
df['YEAR'] = pd.DatetimeIndex(df[' DATE1']).year
df=df.sort_values(by=["SYMBOL","MONTH","YEAR"])
```

## MAKING A NEW COLUMN NAMED CHANGE_RATE TO STORE THE RATE OF CHANGE BETWEEN PREV_CLOSE AND CLOSE_PRICE WHICH WILL HELP US tO DETERMINE BEST AND WORST PERFORMERS

```python
df["CHANGE_RATE"]=(((df[" CLOSE_PRICE"]-df[" PREV_CLOSE"])/df[" PREV_CLOSE"])*100)
```

## REMOVING EXTRA COLUMNS

```python
df=df.drop([" OPEN_PRICE"," HIGH_PRICE"," LOW_PRICE"," LAST_PRICE"," AVG_PRICE"," TTL_TRD_QNTY"," TURNOVER_LACS"," NO_OF_TRADES"," DELIV_QTY"," DELIV_PER"], axis=1)
```

## FILTERING AND SORTING ON THE BASIS OF YEAR AND CHANGE_RATE

```python
df2022 = df[df['YEAR']==2022]
df2022_ascending=df2022.sort_values(by=["CHANGE_RATE"])
df2022_ascending
df2022_descending=df2022.sort_values(by=["CHANGE_RATE"],ascending=False)
df2022_descending


df2021 = df[df['YEAR']==2021]
df2021_ascending=df2021.sort_values(by=["CHANGE_RATE"])
df2021_ascending
df2021_descending=df2021.sort_values(by=["CHANGE_RATE"],ascending=False)
df2021_descending


df2020 = df[df['YEAR']==2020]
df2020_ascending=df2020.sort_values(by=["CHANGE_RATE"])
df2020_ascending
df2020_descending=df2020.sort_values(by=["CHANGE_RATE"],ascending=False)
df2020_descending
```

## USING A FOR LOOP TO GET THE TOP 5 NON PERFORMING STOCKS OF EACH MONTH OF YEAR 2022

```python
for i in range(1,3):
    print("MONTH = {} YEAR=2022".format(i))
    ax = sns.barplot(x='SYMBOL',y='CHANGE_RATE',data=df2022_ascending[df2022_ascending['MONTH']==i].head(5))
    
    plt.show()
```

## USING A FOR LOOP TO GET THE TOP 5 BEST PERFORMING STOCKS OF EACH MONTH OF YEAR 2022

```python
for i in range(1,3):
    print("MONTH = {} YEAR=2022".format(i))
    ax = sns.barplot(x='SYMBOL',y='CHANGE_RATE',data=df2022_descending[df2022_descending['MONTH']==i].head(5))
    plt.show()
```
